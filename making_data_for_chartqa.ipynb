#### creating VLT5 input file

```python
seg_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/seg_data/data/train/tables/"
base_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/data/train/tables/"
image_base_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/data/train/png/"
qa_path = "C:/Users/pande/Desktop/cqa_22_with_id/pmctrain22/"
qa_column = list()
ans_col = list()
qaid_col = list()
img_index = list()
qid = list()
cou = 0
for i in lst_of_folders:
    for j in os.listdir(seg_folder+i):
        file = os.path.splitext(j)[0]
        df = pd.read_csv(seg_folder+i+'/'+j)
        with open(full_json_path+i+'/'+file+'.json','r') as f1,open(qa_path+i+"_qa/"+file+'.json','r') as f2:
            js = json.load(f1)
            js_qa = json.load(f2)
            if 'bar' in js['task1']['output']['chart_type'].split() and 'horizontal' in js['task1']['output']['chart_type'].split():
                chart_type = 'hbar_categorical'
            elif 'bar' in js['task1']['output']['chart_type'].split() and 'vertical' in js['task1']['output']['chart_type'].split():
                chart_type = 'vbar_categorical'
            elif 'line' in js['task1']['output']['chart_type'].split():
                chart_type = 'line'
            elif 'scatter' in js['task1']['output']['chart_type'].split():
                chart_type = 'scatter'
            elif 'box' in js['task1']['output']['chart_type'].split():
                chart_type = 'vbox'
            chart_title,x_label,y_label = str(),str(),str()
            if 'task7' in js.keys():
                if 'chart_title' in js['task7'].keys():
                    chart_title = js['task7']['chart_title']
                if 'x_label' in js['task7'].keys():
                    x_label = js['task7']['x_label'].replace("\n"," ")
                if 'y_label' in js['task7'].keys():
                    y_label = js['task7']['y_label'].replace("\n"," ")
            if y_label!="":
                col = df.columns
                tables = str()
                counter=0
                for cn,cv in df.iteritems():
                    if counter==0:
                        tables+=cn.replace("\n"," ")+" "
                    else:
                        tables+=cn.replace("\n"," ")+" "+y_label+" "
                    for k in list(cv.values):
                        tables+= "| "+str(k).replace("\n"," ")+ " "
                    tables+="& "
                    counter+=1
                tables = tables[:-2]
            else:
                col = df.columns
                tables = str()
                counter=0
                for cn,cv in df.iteritems():
                    tables+=cn.replace("\n"," ")+" "
                    for k in list(cv.values):
                        tables+= "| "+str(k).replace("\n"," ")+ " "
                    tables+="& "
                    counter+=1
                tables = tables[:-2]
            for k in js_qa:
                st = 'Question: '+k['question'].replace("\n"," ")+' '+'Table: '+tables+'Chart Type: '+chart_type
                if chart_title!="":
                    st+=" Title: "+chart_title
                if x_label!="":
                    st+=" x_axis_title: "+x_label
                if y_label!="":
                    st+=" y_axis_title: "+y_label
                qa_column+=["'"+str(st)+"'"]
                ans_col.append(str(k['answer']).replace("\n"," "))
                qaid_col.append(k['qa_id'])
                img_index.append(file)
                cou+=1
                qid.append(cou)
df_final = pd.DataFrame()
df_final['Input'] = qa_column
df_final['Output'] = ans_col
df_final['Image Index'] = img_index
df_final['Question ID'] = qid
df_final['qaid'] = qaid_col
df_final.to_csv("C:/Users/pande/Desktop/cqa_22_with_id/pmctrain22/data_train.csv", sep=',',index = False, encoding='utf-8')
```

```python
seg_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/seg_data/data/val/tables/"
base_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/data/val/tables/"
image_base_folder = "C:/Users/pande/Desktop/Benchmarking_chart_qa/data/val/png/"
qa_path = "C:/Users/pande/Desktop/cqa_22_with_id/pmctrain22/"
qa_column = list()
ans_col = list()
qaid_col = list()
img_index = list()
qid = list()
cou = 0
for i in lst_of_folders:
    for j in os.listdir(seg_folder+i):
        file = os.path.splitext(j)[0]
        df = pd.read_csv(seg_folder+i+'/'+j)
        with open(full_json_path+i+'/'+file+'.json','r') as f1,open(qa_path+i+"_qa/"+file+'.json','r') as f2:
            js = json.load(f1)
            js_qa = json.load(f2)
            if 'bar' in js['task1']['output']['chart_type'].split() and 'horizontal' in js['task1']['output']['chart_type'].split():
                chart_type = 'hbar_categorical'
            elif 'bar' in js['task1']['output']['chart_type'].split() and 'vertical' in js['task1']['output']['chart_type'].split():
                chart_type = 'vbar_categorical'
            elif 'line' in js['task1']['output']['chart_type'].split():
                chart_type = 'line'
            elif 'scatter' in js['task1']['output']['chart_type'].split():
                chart_type = 'scatter'
            elif 'box' in js['task1']['output']['chart_type'].split():
                chart_type = 'vbox'
            chart_title,x_label,y_label = str(),str(),str()
            if 'task7' in js.keys():
                if 'chart_title' in js['task7'].keys():
                    chart_title = js['task7']['chart_title']
                if 'x_label' in js['task7'].keys():
                    x_label = js['task7']['x_label'].replace("\n"," ")
                if 'y_label' in js['task7'].keys():
                    y_label = js['task7']['y_label'].replace("\n"," ")
            if y_label!="":
                col = df.columns
                tables = str()
                counter=0
                for cn,cv in df.iteritems():
                    if counter==0:
                        tables+=cn.replace("\n"," ")+" "
                    else:
                        tables+=cn.replace("\n"," ")+" "+y_label+" "
                    for k in list(cv.values):
                        tables+= "| "+str(k).replace("\n"," ")+ " "
                    tables+="& "
                    counter+=1
                tables = tables[:-2]
            else:
                col = df.columns
                tables = str()
                counter=0
                for cn,cv in df.iteritems():
                    tables+=cn.replace("\n"," ")+" "
                    for k in list(cv.values):
                        tables+= "| "+str(k).replace("\n"," ")+ " "
                    tables+="& "
                    counter+=1
                tables = tables[:-2]
            for k in js_qa:
                st = 'Question: '+k['question'].replace("\n"," ")+' '+'Table: '+tables+'Chart Type: '+chart_type
                if chart_title!="":
                    st+=" Title: "+chart_title
                if x_label!="":
                    st+=" x_axis_title: "+x_label
                if y_label!="":
                    st+=" y_axis_title: "+y_label
                qa_column+=["'"+str(st)+"'"]
                ans_col.append(str(k['answer']).replace("\n"," "))
#                 qaid_col.append(k['qa_id'])
                img_index.append(file)
                cou+=1
                qid.append(cou)
df_final = pd.DataFrame()
df_final['Input'] = qa_column
df_final['Output'] = ans_col
df_final['Image Index'] = img_index
df_final['Question ID'] = qid
# df_final['qaid'] = qaid_col
df_final.to_csv("C:/Users/pande/Desktop/cqa_22_with_id/pmctrain22/data_val.csv", sep=',',index = False, encoding='utf-8')
```
