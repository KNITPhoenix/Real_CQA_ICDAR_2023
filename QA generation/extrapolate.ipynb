```python
import json
import matplotlib.pyplot as plt
```

```python
with open("PMC1193525___pgen.0010015.g006.json") as file:
    data = json.load(file)
```

```python
x_cors = []
y_cors = []
for point in data["task6"]["output"]["data series"][2]["data"]:
    x_cors.append(point['x'])
    y_cors.append(point['y'])
```

```python
print(x_cors)
print(y_cors)
```

```python
def interpolate(x_cors,y_cors,x):
    for i in range(len(x_cors)):
        if x <= x_cors[i]:
            m = (y_cors[i] - y_cors[i-1])/(x_cors[i] - x_cors[i-1])
            return m * (x - x_cors[i]) +  y_cors[i]
    raise Exception("The provided x coordinate is outside the range of the data")
```

```python
x = 5
y = interpolate(x_cors,y_cors,x)
plt.plot(x_cors,y_cors)
plt.plot(x,y,'o')
y
```
